```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
taxis = sns.load_dataset("taxis").head(300)
taxis["tip_bool"] = taxis["tip"] != 0
taxis["dis_h_5km"] = taxis["distance"] > 5
taxis
# taxi zakaz
```

```python
sns.scatterplot(x="fare", y="tip", data=taxis, hue="passengers")
plt.show()
# xizmat narxiga nisbatan choychaqa, odamlar soniga qarab ajratilgan
# bunda taxi narxi qancha qimmat bosa choychaqa koproq bolish extimolligi kattaroq
```

```python
sns.countplot(x="payment", data=taxis, hue="tip_bool")
# naxt tolov qiladiganlar choy-chaqa qoldirmagan ekan
```

```python
sns.relplot(x="total", y="tip", data=taxis, kind="scatter", col="dis_h_5km")
plt.show()
# 5 km dan kop yoki kamga ajratilganda
```

```python
sns.relplot(x="total", y="tip", data=taxis, kind="scatter", size="passengers", hue="passengers")
plt.show()
# birinchi scatter qoshimcha funksiyalari
```

```python
sns.catplot(x="dis_h_5km", y="tip", data=taxis, kind="box")
plt.show()
# minimumlari bir xil 5 km dan kam zakazlarda bita outliner
# ikkinchisni medianasi birinchisini maks bilan teng
```

```python
sns.catplot(x="dropoff_borough", y="fare", data=taxis, kind="point")
plt.show()
# zakazlar narxi har bir shahar uchun
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("phone_price.csv")
df = df.set_index("Unnamed: 0")
a = df["company"].isin(['Coolpad', 'IQOO', 'Itel', 'Gionee', 'LG', 'itel', 'Tecno', 'Lenovo', 'Nothing', 'Lava', 'POCO', 'Asus', 'Google', 'Oppo'])
for i in range(len(a)):
    if a[i]:
        a[i] = False
    else:
        a[i] = True
df = df[a]
df
```

```python
df["company"].value_counts()
```

```python
fig, ax = plt.subplots()
ax.hist(df['company'], bins=11)
plt.xticks(rotation=90)
plt.show()
# telefon sonlari
```

```python
sns.histplot(df['company'])
plt.xticks(rotation=90)
plt.show()
# telefon sonlari part2
```

```python
sns.displot(df["company"], kind='ecdf')
plt.xticks(rotation=90)
plt.show()
# telefon sonlari part3
```

```python
sns.lmplot(data=df, y='Rating', x='Spec_score')
plt.show()
# spec scorega nisbatan reyting, Spec_score qancha ko'p bo'lsa reyting baland bo'lish ehtimoli ko'proq
```

```python
sns.displot(df['fast_charging'])
plt.xticks(rotation=90)
plt.show()
```

```python
sns.relplot(x="Spec_score", y='Rating', data=df, kind='line')
plt.show()
# rating spec_score ganisbatan
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("summer2016.csv")
df
```

```python
df = df.set_index("Unnamed: 0")
```

```python
uzb_boxing = df[(df["NOC"] == "UZB") & (df["Sport"] == "Boxing")]
fra_boxing = df[(df["NOC"] == "FRA") & (df["Sport"] == "Boxing")]
```

# boxing

```python
fig, ax = plt.subplots()
ax.bar("Uzb",uzb_boxing["Height"].mean())
ax.bar("Fra",fra_boxing["Height"].mean())
ax.set_ylabel("Height (cm)")
ax.set_title("Title")
plt.show()
# bunda korishimiz mumkin fransiyalik bokschilarning o'rtacha bo'ylari uzbga qaraganda 6-7 sm ga uzunroq
```

```python
uzb_boxing["Age"].median()
```

```python
fig, ax = plt.subplots()
ax.bar("Uzb",uzb_boxing["Age"].median())
ax.bar("Fra",fra_boxing["Age"].median())
plt.show()
# yoshlarida ham fransida kattaroq
```

```python
men_box = df[(df["Sex"] == "M") & (df["Sport"] == "Boxing")]
women_box = df[(df["Sex"] == "F") & (df["Sport"] == "Boxing")]
```

```python
fig, ax = plt.subplots()
ax.bar("Men", men_box["Age"].mean())
ax.bar("Women",women_box["Age"].mean())
plt.show()
# ayol bokschilar qariroq ekan
```

```python
team_medals = df.groupby(by="Team")["Medal"].count().sort_values(ascending=False).to_frame()
team_medals = team_medals.head(10)
```

```python
fig, ax = plt.subplots()
ax.bar(team_medals.index, team_medals["Medal"])
ax.set_xticklabels(team_medals.index, rotation=45)
plt.show()
# eng ko'p medal olgan 10 ta team
```

```python
fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4, 5, 6], uzb_boxing["Weight"].sort_values().head(6), marker="o", label="Uzb")
ax.plot([1, 2, 3, 4, 5, 6], fra_boxing["Weight"].sort_values().head(6), marker="o", label="Fra")
ax.set_ylabel("Weight (kg)")
ax.legend()
plt.show()
# fransiya va uzb bokschilarining ogirligi
```

```python
mean_height = df[(df["Age"] >= 18)&(df["Age"] <= 38)].groupby(by="Age")["Height"].mean().to_frame()
```

```python
fig, ax = plt.subplots()
ax.plot(mean_height.index, mean_height["Height"], marker="v")
ax.set_xlabel("Age Y")
ax.set_ylabel("Height (cm)")
plt.show()
# 18 dan 39 gacha bolgan sportchilarni ortacha boyi
```

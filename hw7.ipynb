```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("seattle_weather.csv")
```

```python
stations = df["STATION"].unique()
st1, st2 = stations[0], stations[1]
data1 = df[df["STATION"] == st1]
data2 = df[df["STATION"] == st2]
tavg = "MLY-TAVG-NORMAL"
```

```python
fig, ax = plt.subplots(2, 2)
ax[0, 0].scatter(df["MLY-PRCP-NORMAL"], df["MLY-TAVG-NORMAL"], c=df["MLY-TAVG-NORMAL"])
ax[1, 0].bar(st1, data1[tavg].mean(), yerr=data1[tavg].std())
ax[1, 0].bar(st2, data2[tavg].mean(), yerr=data2[tavg].std())
ax[0, 1].boxplot(df["MLY-PRCP-NORMAL"])
ax[1, 1].plot(data1["DATE"], data1[tavg], marker="o", label="st1")
ax[1, 1].plot(data2["DATE"], data2[tavg], label="st2")
ax[1, 1].legend()
plt.show()

# 1) temperatura o'zgarishi yomg'irga nisbatan
# 2) ....
# 3) 2 ta stansiyaning o'rtacha yillik temp boyicha ma'lumotlari stadard deviation bilan
# 4) shu ma'lumotlar lineplot ko'rinishida
```
